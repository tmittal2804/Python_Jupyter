1. Write a program to count and display the number of capital letters in a given string.

#### Program Code and Output:

```python
string = input("Enter a string: ")
count = 0

for ch in string:
    if ch.isupper():
        count += 1

print("Entered string:", string)
print("Number of capital letters:", count)
```

2. Count total number of vowels in a given string.

#### Program Code and Output:

```python
string = input("Enter a string: ")
vowels = "aeiouAEIOU"
count = 0

for ch in string:
    if ch in vowels:
        count += 1

print("Entered string:", string)
print("Total number of vowels:", count)
```

4. WAP to enter a string and a substring. You have to print the number of times that the substring occurs in the given string. String traversal will take place from left to right, not from right to left.

#### Program Code and Output:

```python
string = input("Enter main string: ")
substring = input("Enter substring: ")

count = 0

for i in range(len(string) - len(substring) + 1):
    if string[i:i+len(substring)] == substring:
        count += 1

print("Main string:", string)
print("Substring:", substring)
print("Number of occurrences:", count)
```

5. Given a string containing both upper and lower case alphabets. Write a Python program to count the number of occurrences of each alphabet (case insensitive) and display the same.

#### Program Code and Output:

```python
string = input("Enter a string: ").upper()

freq = {}

for ch in string:
    if ch.isalpha():
        freq[ch] = freq.get(ch, 0) + 1

for key in sorted(freq):
    print(freq[key], key)
```

6. Program to count number of unique words in a given sentence using sets.

#### Program Code and Output:

```python
sentence = input("Enter a sentence: ")

words = sentence.split()
unique_words = set(words)

print("Entered sentence:", sentence)
print("Number of unique words:", len(unique_words))
```

7.	Create 2 sets s1 and s2 of n fruits each by taking input from user and find:
a)	Fruits which are in both sets s1 and s2
b)	Fruits only in s1 but not in s2
c)	Count of all fruits from s1 and s2

#### Program Code and Output:

```python
n = int(input("Enter number of fruits in each set: "))

s1 = set()
s2 = set()

print("\nEnter fruits for Set 1:")
for i in range(n):
    fruit = input(f"Fruit {i+1}: ")
    s1.add(fruit)

print("\nEnter fruits for Set 2:")
for i in range(n):
    fruit = input(f"Fruit {i+1}: ")
    s2.add(fruit)

print("\nFruits in Set 1:", sorted(s1))
print("Fruits in Set 2:", sorted(s2))

common_fruits = s1.intersection(s2)
only_s1 = s1.difference(s2)
all_fruits = s1.union(s2)
total_count = len(all_fruits)

print("\nFruits in both sets:", sorted(common_fruits))
print("Fruits only in Set 1:", sorted(only_s1))
print("Total count of unique fruits from both sets:", total_count)
```

8.	Take two sets and apply various set operations on them :
S1 = {Red ,yellow, orange , blue }
S2 = {violet, blue , purple}

#### Program Code and Output:

```python
S1 = {"Red", "Yellow", "Orange", "Blue"}
S2 = {"Violet", "Blue", "Purple"}

print("Set S1:", S1)
print("Set S2:", S2)

print("\nUnion (S1 u S2):", S1.union(S2))
print("Intersection (S1 ∩ S2):", S1.intersection(S2))

print("Difference (S1 - S2):", S1.difference(S2))
print("Difference (S2 - S1):", S2.difference(S1))

print("Symmetric Difference (S1 △ S2):", S1.symmetric_difference(S2))
```

### Observation:

## Experiment-5: Lists, Tuples and Dictionary

### Aim:
To implement Python programs using list, tuple, and dictionary data structures and display user input along with required outputs.

1. To implement Python programs using list, tuple, and dictionary data structures and display user input along with required outputs.

#### Program Code and Output:

```python
n = int(input("Enter number of values: "))
print("You entered n =", n)

count = [0, 0, 0, 0]
entered_values = []

for i in range(n):
    value = int(input("Enter value (0-3): "))
    entered_values.append(value)
    
    if 0 <= value <= 3:
        count[value] += 1
    else:
        print("Invalid input!")

print("Values entered by user:", entered_values)

for i in range(4):
    print(f"{i} occurred {count[i]} times")
```

2.	Create a tuple to store n numeric values and find average of all values.

#### Program Code and Output:

```python
n = int(input("Enter number of values: "))
print("You entered n =", n)

values = []

for i in range(n):
    num = float(input("Enter number: "))
    values.append(num)

t = tuple(values)

print("Values entered:", values)
print("Tuple created:", t)

average = sum(t) / len(t)
print("Average of tuple values:", average)
```

4.	Create a dictionary of n persons where key is name and value is city. 
a) Display all names
b) Display all city names
c) Display student name and city of all students.
d) Count number of students in each city.

#### Program Code and Output:

```python
n = int(input("Enter number of persons: "))
print("You entered n =", n)

students = {}

for i in range(n):
    name = input("Enter name: ")
    city = input("Enter city: ")
    students[name] = city

print("\nDictionary created:", students)

print("\nNames:")
for name in students.keys():
    print(name)

print("\nCities:")
for city in students.values():
    print(city)

print("\nStudent Details:")
for name, city in students.items():
    print(name, "lives in", city)

city_count = {}
for city in students.values():
    city_count[city] = city_count.get(city, 0) + 1

print("\nStudents in each city:")
for city, count in city_count.items():
    print(city, ":", count)
```

5.	Store details of n movies in a dictionary by taking input from the user. Each movie must store details like name,  year, director name, production cost, collection made (earning) & perform the following :-
a)	print all movie details
b)	display name of movies released before 2015
c)	print movies that made a profit.
d)	print movies directed by a particular director.

#### Program Code and Output:

```python
n = int(input("Enter number of movies: "))
print("You entered n =", n)

movies = {}

for i in range(n):
    print("\nEnter details for movie", i+1)
    name = input("Movie Name: ")
    year = int(input("Release Year: "))
    director = input("Director Name: ")
    cost = float(input("Production Cost: "))
    collection = float(input("Collection Made: "))

    movies[name] = {
        "year": year,
        "director": director,
        "cost": cost,
        "collection": collection
    }

print("\nMovies dictionary created:")
for name, details in movies.items():
    print(name, ":", details)

print("\nMovies released before 2015:")
for name, details in movies.items():
    if details["year"] < 2015:
        print(name)

print("\nProfitable Movies:")
for name, details in movies.items():
    if details["collection"] > details["cost"]:
        print(name)

search_director = input("\nEnter director name to search: ")
print("You searched for director:", search_director)

for name, details in movies.items():
    if details["director"] == search_director:
        print(name)
```

6.	Create a contact book where users can store, search, update, and delete contacts. Use dictionary for storing contacts.

#### Program Code and Output:

```python
contacts = {}

while True:
    print("\n1. Add Contact")
    print("2. Search Contact")
    print("3. Exit")

    choice = int(input("Enter choice: "))
    print("You selected:", choice)

    if choice == 1:
        name = input("Enter name: ")
        phone = input("Enter phone: ")
        contacts[name] = phone
        print("Contact added!")
        print("Current contacts:", contacts)

    elif choice == 2:
        name = input("Enter name to search: ")
        print("Phone:", contacts.get(name, "Not found"))

    elif choice == 3:
        print("Exiting...")
        break
```

7.	Create a Todo list Manager where users can add, view, and remove tasks. Use List for storing tasks.

#### Program Code and Output:

```python
tasks = []

while True:
    print("\n1. Add Task")
    print("2. View Tasks")
    print("3. Exit")

    choice = int(input("Enter choice: "))
    print("You selected:", choice)

    if choice == 1:
        task = input("Enter task: ")
        tasks.append(task)
        print("Task added!")
        print("Current Tasks:", tasks)

    elif choice == 2:
        print("Your Tasks:")
        for i, task in enumerate(tasks, start=1):
            print(i, ".", task)

    elif choice == 3:
        print("Exiting...")
        break
```

-  Tuples store fixed data and are useful for calculations like average.
-  Dictionaries store data in key-value pairs for fast searching and updating.
-  Set concept helps in removing duplicate scores while finding runner-up.
