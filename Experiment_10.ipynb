1. Create numpy array to find sum of all elements in an array.

#### Program Code and Output:

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])
total_sum = np.sum(arr)

print("Array:", arr)
print("Sum of all elements:", total_sum)
```

2. Create numpy array of (3,3) dimension. Now find sum of all rows & columns individually. Also find 2nd maximum element in the array. 

#### Program Code and Output:

```python
arr2 = np.array([[5, 8, 2],
                 [1, 9, 7],
                 [4, 6, 3]])

print("Array:\n", arr2)

# Row-wise sum
row_sum = np.sum(arr2, axis=1)
print("Row-wise sum:", row_sum)

# Column-wise sum
col_sum = np.sum(arr2, axis=0)
print("Column-wise sum:", col_sum)

# 2nd maximum element
second_max = np.sort(arr2, axis=None)[-2]
print("Second maximum element:", second_max)
```

3. Perform Matrix multiplication of any 2 n*n matrices.

#### Program Code and Output:

```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

result = np.dot(A, B)

print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Matrix Multiplication Result:\n", result)
```

4.	 Write a Pandas program to get the powers of an array values element-wise. 
Note: First array elements raised to powers from second array
Sample data: {'X':[78,85,96,80,86], 'Y':[84,94,89,83,86],'Z':[86,97,96,72,83]}
Expected Output:
X  Y  Z
0 78 84 86
1 85 94 97
2 96 89 96
3 80 83 72
4 86 86 83

#### Program Code and Output:

```python
import pandas as pd

data = {
    'X': [78, 85, 96, 80, 86],
    'Y': [84, 94, 89, 83, 86],
    'Z': [86, 97, 96, 72, 83]
}

df = pd.DataFrame(data)

# Raising values element-wise (example: X^Y)
result = df['X'] ** df['Y']

print("Original DataFrame:\n", df)
print("Element-wise Power (X^Y):\n", result)
```

7. Create a program to demonstrate different visual forms using Matplotlib.

#### Program Code and Output:

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

# Line Plot
plt.figure()
plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

# Bar Chart
plt.figure()
plt.bar(x, y)
plt.title("Bar Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

# Scatter Plot
plt.figure()
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

# Pie Chart
plt.figure()
labels = ['A', 'B', 'C', 'D', 'E']
plt.pie(y, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```
