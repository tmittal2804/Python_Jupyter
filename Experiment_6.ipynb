1. Write a Python function to find the maximum and minimum numbers from a sequence of numbers.  (Note: Do not use built-in functions.)

#### Program Code and Output:

```python
def find_max_min(numbers):
    maximum = numbers[0]
    minimum = numbers[0]

    for num in numbers:
        if num > maximum:
            maximum = num
        if num < minimum:
            minimum = num

    return maximum, minimum


nums = list(map(int, input("Enter numbers separated by space: ").split()))
print("You entered:", nums)

result = find_max_min(nums)
print("Maximum and Minimum:", result)
```

2.	Write a Python function that takes a positive integer and returns the sum of the cube of all the positive integers smaller than the specified number. 

#### Program Code and Output:

```python
def sum_of_cubes(n):
    total = 0
    for i in range(1, n):
        total += i**3
    return total

n = int(input("Enter a positive integer: "))
print("You entered:", n)

print("Sum of cubes:", sum_of_cubes(n))
```

3.	Write a Python function to print 1 to n using recursion. (Note: Do not use loop)

#### Program Code and Output:

```python
def print_numbers(n):
    if n == 0:
        return
    print_numbers(n-1)
    print(n)

n = int(input("Enter value of n: "))
print("You entered:", n)

print("Numbers from 1 to n:")
print_numbers(n)
```

4.	Write a recursive function to print Fibonacci series upto n terms.

#### Program Code and Output:

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

def print_fibonacci(n):
    for i in range(n):
        print(fibonacci(i), end=" ")

n = int(input("Enter number of terms: "))
print("You entered:", n)

print("Fibonacci Series:")
print_fibonacci(n)
```

5.	Write a lambda function to find volume of cone.

#### Program Code and Output:

```python
import math

volume = lambda r, h: (1/3) * math.pi * r * r * h

r = float(input("Enter radius: "))
h = float(input("Enter height: "))

print("Radius:", r)
print("Height:", h)

print("Volume of Cone:", volume(r, h))
```

6.	Write a lambda function which gives tuple of max and min from a list.

#### Program Code and Output:

```python
lst = list(map(int, input("Enter numbers separated by space: ").split()))
print("You entered:", lst)

max_min = lambda x: (max(x), min(x))

print("Output:", max_min(lst))
```

7.	Write functions to explain mentioned concepts:
    a) Keyword argument
    b) Default argument
    c) Variable length argument

#### Program Code and Output:

```python
# Keyword Argument
def student(name, age):
    print("Name:", name)
    print("Age:", age)

name = input("Enter name: ")
age = int(input("Enter age: "))

student(age=age, name=name)
```

```python
# Default Argument
def greet(name="People"):
    print("Hello", name)

name = input("Enter name (leave blank for default): ")

if name == "":
    greet()
else:
    greet(name)
```

```python
# Variable Length Argument
def add_numbers(*numbers):
    total = 0
    for num in numbers:
        total += num
    print("Sum:", total)

nums = list(map(int, input("Enter numbers separated by space: ").split()))
print("You entered:", nums)

add_numbers(*nums)
```

8.	Write a program to check whether all the values in a dictionary are same or not using lambda function.

#### Program Code and Output:

```python
n = int(input("Enter number of items in dictionary: "))
d = {}

for i in range(n):
    key = input("Enter key: ")
    value = input("Enter value: ")
    d[key] = value

print("You entered:", d)

check_same = lambda x: len(set(x.values())) == 1

print("All values same?", check_same(d))
```

9.	Write a program to create two lists and generate a dictionary with keys from list1 and values from list2.

#### Program Code and Output:

```python
list1 = input("Enter keys separated by space: ").split()
list2 = input("Enter values separated by space: ").split()

print("You entered keys:", list1)
print("You entered values:", list2)

dictionary = dict(zip(list1, list2))
print("Generated Dictionary:", dictionary)
```
