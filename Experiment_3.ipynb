1.	Find a factorial of given number.

#### Program Code and Output:

```python
num = int(input("Enter a number: "))

print("Entered number:", num)

fact = 1
for i in range(1, num + 1):
    fact *= i

print("Factorial of", num, "is", fact)
```

2. Find whether the given number is Armstrong number.

#### Program Code and Output:

```python
n = int(input("Enter number: "))
print("Entered number:", n)

temp = n
digits = len(str(n))
sum = 0

while temp > 0:
    d = temp % 10
    sum += d ** digits
    temp //= 10

if sum == n:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

3.	Print Fibonacci series up to given term.

#### Program Code and Output:

```python
n = int(input("Enter number of terms: "))
print("Number of terms entered:", n)

a, b = 0, 1

print("Fibonacci Series:")
for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

4.	Write a program to find if given number is prime number or not.

#### Program Code and Output:

```python
n = int(input("Enter number: "))
print("Entered number:", n)

if n > 1:
    for i in range(2, n):
        if n % i == 0:
            print(n, "is Not Prime")
            break
    else:
        print(n, "is a Prime Number")
else:
    print(n, "is Not Prime")
```

5.	Check whether given number is palindrome or not.

#### Program Code and  Output:

```python
n = int(input("Enter number: "))
print("Entered number:", n)

rev = int(str(n)[::-1])

if n == rev:
    print(n, "is a Palindrome")
else:
    print(n, "is Not a Palindrome")
```

6.	Write a program to print sum of digits.

#### Program Code and Output:

```python
n = int(input("Enter number: "))
print("Entered number:", n)

temp = n  # store original number for display
sum_digits = 0

while temp > 0:
    sum_digits += temp % 10
    temp //= 10

print("Sum of digits of", n, "is", sum_digits)
```

7.	Count and print all numbers divisible by 5 or 7 between 1 to 100.

#### Program Code and Output:

```python
count = 0

for i in range(1, 101):
    if i % 5 == 0 or i % 7 == 0:
        print(i, end=" ")
        count += 1

print("\nCount =", count)
```

10.	Write a program to print the following pattern
123454321
1234 *4321
123  * * 321
12   * * *  21
1    * * * *   1

#### Program Code and Output:

```python
n = int(input("Enter number of rows: "))
print("Number of rows entered:", n)
print("Pattern:")

for i in range(n, 0, -1):
    # Print left numbers
    for j in range(1, i + 1):
        print(j, end="")

    # Print spaces and stars
    if i < n:
        print(" " * (n - i), end="")
        for k in range(n - i):
            print("*", end=" ")
    
    # Print right numbers
    for j in range(i, 0, -1):
        print(j, end="")
    
    print()  # Move to next line
```

11.	Write a program to print the sum of the following series
1+ ½ + 1/3 + ¼ +….+1/n

#### Program Code and Output:

```python
n = int(input("Enter the value of n: "))
print("Value of n entered:", n)

sum_series = 0.0

for i in range(1, n + 1):
    sum_series += 1 / i

print("Sum of the series 1 + 1/2 + 1/3 + ... + 1/", n, "is", sum_series)
```
