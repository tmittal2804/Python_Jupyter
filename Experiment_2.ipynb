## Experiment-2: Conditional Statements

### Aim:
To write and execute Python programs that demonstrate the use of conditional statements (if, elif, else) for solving basic real-world problems such as checking divisibility, finding greatest numbers, validating leap years, determining roots of quadratic equations, computing next calendar date, and generating a student grade sheet based on CGPA.

1) Check whether the given number is divisible by 3 and 5 both.

#### Program Code and Output:

```python
num = int(input("Enter a number: "))

if num % 3 == 0 and num % 5 == 0:
    print("Number is divisible by both 3 and 5")
else:
    print("Number is NOT divisible by both 3 and 5")
```

2) Check whether a given number is multiple of five or not.

#### Program Code and Output:

```python
num = int(input("Enter a number: "))

if num % 5 == 0:
    print("Number is a multiple of 5")
else:
    print("Number is NOT a multiple of 5")
```

3) Find the greatest among the two numbers. If numbers are equal than print “numbers are equal”.

#### Program Code and Output:

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("First number:", a)
print("Second number:", b)

if a > b:
    print("Greatest number is:", a)
elif b > a:
    print("Greatest number is:", b)
else:
    print("Numbers are equal")
```

4) Find the greatest among three numbers assuming no two values are same.

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

print("First number:", a)
print("Second number:", b)
print("Third number:", c)

if a > b and a > c:
    print("Greatest number is:", a)
elif b > a and b > c:
    print("Greatest number is:", b)
else:
    print("Greatest number is:", c)
```

5) Check whether the quadratic equation has real roots or imaginary roots. Display the roots.

#### Program Code and Output:

```python
import math

a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))

print("Value of a:", a)
print("Value of b:", b)
print("Value of c:", c)

d = b*b - 4*a*c

if d > 0:
    r1 = (-b + math.sqrt(d)) / (2*a)
    r2 = (-b - math.sqrt(d)) / (2*a)
    print("Real and distinct roots:", r1, r2)
elif d == 0:
    r = -b / (2*a)
    print("Real and equal roots:", r)
else:
    real = -b / (2*a)
    imag = math.sqrt(-d) / (2*a)
    print("Imaginary roots:")
    print(real, "+", imag, "i")
    print(real, "-", imag, "i")
```

6) Find whether a given year is a leap year or not.

#### Program Code and Output:

```python
year = int(input("Enter year: "))

print("Entered year:", year)

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

7) Write a program which takes any date as input and display next date of the calendar.

#### Program Code and Output:

```python
day = int(input("Enter day: "))
month = int(input("Enter month: "))
year = int(input("Enter year: "))

print("\nEntered Date:")
print("Day:", day)
print("Month:", month)
print("Year:", year)

days = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

# Leap year check
if year % 400 == 0 or (year % 4 == 0 and year % 100 != 0):
    days[1] = 29

# Check for valid date
if month < 1 or month > 12 or day < 1 or day > days[month - 1]:
    print("\nInvalid date entered")
else:
    day += 1

    if day > days[month - 1]:
        day = 1
        month += 1

    if month > 12:
        month = 1
        year += 1

    print("\nPredicted (Next) Date:")
    print("Day:", day)
    print("Month:", month)
    print("Year:", year)
```

8) Print the grade sheet of a student for the given range of cgpa. Scan marks of five subjects and calculate the percentage.
CGPA=percentage/10
CGPA range:
0 to 3.4 -> F
3.5 to 5.0->C+
5.1 to 6->B
6.1 to 7-> B+
7.1 to 8-> A
8.1 to 9->A+
9.1 to 10-> O (Outstanding)

#### Program Code and Output:

```python
name = input("Enter Name: ")
roll = input("Enter Roll Number: ")
sapid = input("Enter SAP ID: ")
sem = input("Enter Semester: ")
course = input("Enter Course: ")

print("\nEnter marks of five subjects:")
dsa = int(input("Data Structures and Algorithms: "))
python = int(input("Python: "))
ai_ml = int(input("Elements of AI/ML: "))
de = int(input("Digital Electronics: "))
aem2 = int(input("Advanced Engineering Mathematics-2: "))

total = dsa + python + ai_ml + de + aem2
percentage = total / 5
cgpa = percentage / 10

# Grade Calculation
if cgpa <= 3.4:
    grade = "F"
elif cgpa <= 5.0:
    grade = "C+"
elif cgpa <= 6.0:
    grade = "B"
elif cgpa <= 7.0:
    grade = "B+"
elif cgpa <= 8.0:
    grade = "A"
elif cgpa <= 9.0:
    grade = "A+"
else:
    grade = "O (Outstanding)"

# Display Grade Sheet
print("\n----- GRADE SHEET -----")
print("Name:", name)
print("Roll Number:", roll, "\tSAPID:", sapid)
print("Sem:", sem, "\t\tCourse:", course)

print("\nSubject\t\t\tMarks")
print("Data Structures & Algorithms:\t", dsa)
print("Python:\t\t\t", python)
print("Elements of AI/ML:\t\t", ai_ml)
print("Digital Electronics:\t\t", de)
print("Advanced Engg. Maths-2:\t\t", aem2)

print("\nPercentage:", percentage, "%")
print("CGPA:", round(cgpa, 2))
print("Grade:", grade)
```
