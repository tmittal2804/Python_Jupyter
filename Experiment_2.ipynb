4) Write a program to compute the length of the hypotenuse (c) of a right triangle using Pythagoras theorem. 

```python
foot = float(input("Enter foot: "))
perpendicular = float(input("Enter perpendicular: "))

hypotenuse = (foot*foot + perpendicular*perpendicular) ** 0.5

print("Foot:", foot)
print("Perpendicular:", perpendicular)
print("Hypotenuse:", hypotenuse)
```

5) Write a program to find simple interest

```python
p = float(input("Enter principal: "))
r = float(input("Enter rate: "))
t = float(input("Enter time: "))

si = (p * r * t) / 100

print("Principal:", p)
print("Rate:", r)
print("Time:", t)
print("Simple Interest:", si)
```

6) Write a program to find area of triangle when length of sides are given.

```python
a = float(input("Enter side a: "))
b = float(input("Enter side b: "))
c = float(input("Enter side c: "))

s = (a + b + c) / 2
area = (s * (s - a) * (s - b) * (s - c)) ** 0.5

print("Side a:", a)
print("Side b:", b)
print("Side c:", c)
print("Area of triangle:", area)
```

7) Write a program to convert given seconds into hours, minutes and remaining seconds.

```python
seconds = int(input("Enter seconds: "))

hours = seconds // 3600
minutes = (seconds % 3600) // 60
sec = seconds % 60

print("Hours:", hours)
print("Minutes:", minutes)
print("Seconds:", sec)
```

8) Write a program to swap two numbers without taking additional variable.

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("Before swapping:")
print("a =", a)
print("b =", b)

a = a + b
b = a - b
a = a - b

print("After swapping:")
print("a =", a)
print("b =", b)
```

9) Write a program to find sum of first n natural numbers.

```python
n = int(input("Enter value of n: "))

sum_n = n * (n + 1) // 2

print("n =", n)
print("Sum of first", n, "natural numbers:", sum_n)
```

10) Write a program to print truth table for bitwise operators( & , | and ^ operators)

```python
print("A B  A&B  A|B  A^B")

for A in [0, 1]:
    for B in [0, 1]:
        print(A, B, " ", A & B, "   ", A | B, "   ", A ^ B)
```

11) Write a program to find left shift and right shift values of a given number.

```python
n = int(input("Enter a number: "))

left_shift = n << 1
right_shift = n >> 1

print("Number:", n)
print("Left shift (n << 1):", left_shift)
print("Right shift (n >> 1):", right_shift)
```

12) Using membership operator find whether a given number is in sequence (10,20,56,78,89)

```python
sequence = (10, 20, 56, 78, 89)

num = int(input("Enter a number: "))

print("Sequence:", sequence)
print("Entered number:", num)

if num in sequence:
    print("The number is present in the sequence")
else:
    print("The number is not present in the sequence")
```

13) Using membership operator find whether a given character is in a string.

```python
s = input("Enter a string: ")
ch = input("Enter a character: ")

print("String:", s)
print("Character:", ch)

if ch in s:
    print("The character is present in the string")
else:
    print("The character is not present in the string")
```
