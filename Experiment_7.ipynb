1.	Add few names, one name in each row, in “name.txt file”.
a.	Count no of names
b.	Count all names starting with vowel
c.	Find longest name

#### Program Code and Output:

```python
with open("name.txt", "r", encoding="utf-8") as f:
    names = [line.strip() for line in f if line.strip()]

print("Total names:", len(names))

vowels = ('A','E','I','O','U','a','e','i','o','u')
vowel_names = [name for name in names if name.startswith(vowels)]
print("Names starting with vowel:", len(vowel_names))

if names:
    print("Longest name:", max(names, key=len))
else:
    print("File is empty")
```

2.	Store integers in a file.
a.	Find the max number
b.	Find average of all numbers
c.	Count number of numbers greater than 100

#### Program Code and Output:

```python
# create file
with open("numbers.txt", "w") as f:
    f.write("10 50 120 200 30 150")

# now read it
with open("numbers.txt", "r") as f:
    nums = [int(x) for x in f.read().split()]

print("Maximum:", max(nums))
print("Average:", sum(nums)/len(nums))
print("Numbers > 100:", sum(1 for x in nums if x > 100))
```

3.	Assume a file city.txt with details of 5 cities in given format (cityname population(in lakhs) area(in sq KM) ):
Example:
Dehradun 5.78 308.20
Delhi 190 1484

#### Program Code and Output:

```python
with open("city.txt", "w") as f:
    f.write("Dehradun 5.78 308.20\n")
    f.write("Delhi 190 1484\n")
    f.write("Mumbai 124 603.4\n")
    f.write("Chandigarh 11 114\n")
    f.write("Jaipur 30 484.6\n")
cities = []

with open("city.txt", "r") as f:
    for line in f:
        name, pop, area = line.split()
        cities.append((name, float(pop), float(area)))

# a) Display all city details
print("All city details:")
for c in cities:
    print(c)

# b) Cities with population > 10 lakhs
print("\nCities with population > 10 lakhs:")
for c in cities:
    if c[1] > 10:
        print(c[0])

# c) Sum of all areas
total_area = sum(c[2] for c in cities)
print("\nTotal area:", total_area)
```

4. Input two values from user where the first line contains N, the number of test cases. The next N lines contain the space separated values of a and b. Perform integer division and print a/b. Handle exception in case of ZeroDivisionError or ValueError. 
Sample input
1 0
2 $
3 1 
Sample Output :
Error Code: integer division or modulo by zero 
Error Code: invalid literal for int() with base 10: '$' 3

#### Program Code and Output:

```python
n = int(input())

for _ in range(n):
    try:
        a, b = input().split()
        print(int(a) // int(b))
        
    except ZeroDivisionError as e:
        print("Error Code:", e)
        
    except ValueError as e:
        print("Error Code:", e)
```

5.	Create multiple suitable exceptions for a file handling program. 

#### Program Code and Output:

```python
# Custom Exceptions
class FileNotFoundCustom(Exception):
    pass

class FileEmptyError(Exception):
    pass

class InvalidDataError(Exception):
    pass


# Step 1: Create and write into file
with open("data.txt", "w") as f:
    f.write("10 20 30 abc")   # change content to test different cases


try:
    filename = "data.txt"

    # Step 2: Open file
    try:
        f = open(filename, "r")
    except FileNotFoundError:
        raise FileNotFoundCustom("File not found!")

    content = f.read()

    # Step 3: Check empty file
    if content.strip() == "":
        raise FileEmptyError("File is empty!")

    # Step 4: Validate data (only numbers allowed)
    for x in content.split():
        if not x.isdigit():
            raise InvalidDataError("File contains invalid data!")

    print("File content is valid:", content)

    f.close()

# Step 5: Handle exceptions
except FileNotFoundCustom as e:
    print(e)

except FileEmptyError as e:
    print(e)

except InvalidDataError as e:
    print(e)
```

6.	Write a program to create a counter to show that how many times the program is executed.

#### Program Code and Output:

```python
try:
    # Read previous count
    with open("counter.txt", "r") as f:
        count = int(f.read())
except:
    # If file doesn't exist or empty
    count = 0

# Increment count
count += 1

# Write updated count
with open("counter.txt", "w") as f:
    f.write(str(count))

print("Program executed", count, "times")
```

### Observation:
It is observed that file handling in Python allows persistent storage of data. Even after the program stops, the execution count is maintained and updated correctly each time the program runs.

### Result:
The program successfully counts and displays the number of times it has been executed by updating a value stored in a file.
