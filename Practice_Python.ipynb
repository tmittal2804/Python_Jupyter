```python
# SMART STUDENT UTILITY SYSTEM

# AAcademic  Evaluation
def academic_evaluation():
    print("\n Academic Evaluation ")
    marks = []
    
    for i in range(5):
        m = float(input(f"Enter marks of subject {i+1}: "))
        marks.append(m)
    
    total = sum(marks)
    percentage = total / 5
    cgpa = percentage / 10

    # Grade Assignment
    if cgpa >= 9:
        grade = "A+"
    elif cgpa >= 8:
        grade = "A"
    elif cgpa >= 7:
        grade = "B"
    elif cgpa >= 6:
        grade = "C"
    elif cgpa >= 5:
        grade = "D"
    else:
        grade = "F"

    print("\n----- Grade Sheet -----")
    print("Marks:", marks)
    print(f"Percentage: {percentage:.2f}%")
    print(f"CGPA: {cgpa:.2f}")
    print("Grade:", grade)


# Mathematical Analyser
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n/2)+1):
        if n % i == 0:
            return False
    return True

def is_armstrong(n):
    power = len(str(n))
    sum_val = sum(int(d)**power for d in str(n))
    return sum_val == n

def is_palindrome(n):
    return str(n) == str(n)[::-1]

def factorial(n):
    fact = 1
    for i in range(1, n+1):
        fact *= i
    return fact

def fibonacci(n):
    a, b = 0, 1
    series = []
    for _ in range(n):
        series.append(a)
        a, b = b, a+b
    return series

def mathematical_analyzer():
    print("\n--- Mathematical Analyzer ---")
    n = int(input("Enter a number: "))

    print("Prime:", is_prime(n))
    print("Armstrong:", is_armstrong(n))
    print("Palindrome:", is_palindrome(n))
    print("Factorial:", factorial(n))

    terms = int(input("Enter number of Fibonacci terms: "))
    print("Fibonacci Series:", fibonacci(terms))

    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    c = int(input("Enter third number: "))
    print("Greatest number:", max(a, b, c))

def string_processing():
    print("\n--- String Processing ---")
    text = input("Enter a sentence: ")

    vowels = "aeiouAEIOU"
    vowel_count = sum(1 for ch in text if ch in vowels)
    upper_count = sum(1 for ch in text if ch.isupper())

    substring = input("Enter substring to search: ")
    sub_count = text.count(substring)

    words = text.split()

    print("\nVowels:", vowel_count)
    print("Uppercase letters:", upper_count)
    print("Substring occurrences:", sub_count)

    print("\nWords in new lines:")
    for w in words:
        print(w)

contacts = {}
todo_list = []

def contact_book():
    while True:
        print("\n--- Contact Book ---")
        print("1. Add Contact")
        print("2. Search Contact")
        print("3. Delete Contact")
        print("4. Back")

        choice = input("Enter choice: ")

        if choice == "1":
            name = input("Enter name: ")
            phone = input("Enter phone: ")
            contacts[name] = phone
            print("Contact added!")

        elif choice == "2":
            name = input("Enter name to search: ")
            print("Phone:", contacts.get(name, "Not Found"))

        elif choice == "3":
            name = input("Enter name to delete: ")
            if name in contacts:
                del contacts[name]
                print("Deleted successfully!")
            else:
                print("Contact not found!")

        elif choice == "4":
            break


def todo():
    while True:
        print("\n--- To-Do List ---")
        print("1. Add Task")
        print("2. View Tasks")
        print("3. Remove Task")
        print("4. Back")

        choice = input("Enter choice: ")

        if choice == "1":
            task = input("Enter task: ")
            todo_list.append(task)
            print("Task added!")

        elif choice == "2":
            print("\nTasks:")
            for i, t in enumerate(todo_list, 1):
                print(i, ".", t)

        elif choice == "3":
            num = int(input("Enter task number to remove: "))
            if 0 < num <= len(todo_list):
                removed = todo_list.pop(num-1)
                print("Removed:", removed)
            else:
                print("Invalid choice!")

        elif choice == "4":
            break


# Main Menu
def main():
    while True:
        print("\n====== SMART STUDENT UTILITY SYSTEM ======")
        print("1. Academic Evaluation")
        print("2. Mathematical Analyzer")
        print("3. String Processing")
        print("4. Data Management")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            academic_evaluation()
        elif choice == "2":
            mathematical_analyzer()
        elif choice == "3":
            string_processing()
        elif choice == "4":
            while True:
                print("\n--- Data Management ---")
                print("1. Contact Book")
                print("2. To-Do List")
                print("3. Back")

                sub = input("Enter choice: ")

                if sub == "1":
                    contact_book()
                elif sub == "2":
                    todo()
                elif sub == "3":
                    break

        elif choice == "5":
            print("Exiting program...")
            break
        else:
            print("Invalid choice!")
```
